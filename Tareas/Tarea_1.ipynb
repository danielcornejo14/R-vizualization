```r
library(ggplot2)
job_data <- read.csv("../DataSets/jobs_in_data.csv")
colnames(job_data)
```

```r
ggplot(job_data, aes(x = job_category, y = salary_in_usd/1000)) +
  geom_boxplot() +
  labs(
    x = "Job Category",
    y = "Annual USD Salary / 10000",
  ) +
  theme(
    axis.text.x = element_text(angle = -45, hjust = 0)
  )
```

```r
ggplot(job_data, aes(job_category, fill = work_setting)) +
  geom_bar() +
  labs(
    title = "Data Science: Count by Job Category",
    x = "Job Category",
    y = "Count",
    fill = "Work Setting"
  ) +
  theme(
    axis.text.x = element_text(angle = -45, hjust = 0)
  )
```

```r
ggplot(job_data, aes(x = experience_level, y = salary_in_usd/1000)) +
  geom_violin() +
  labs(
    title = "Salaries by Experience Level",
    x = "Experience Level",
    y = "Annual USD Salary / 10000"
  ) +
  theme(
    axis.text.x = element_text(angle = -45, hjust = 0)
  )
```

```r
ggplot(job_data, aes(experience_level, fill = work_year)) +
  geom_bar() +
  facet_wrap(~work_setting) +
  labs(
    title = "Experience level: Count by Work Setting and Year",
    x = "Experience Level",
    y = "Count"
  )+
  theme(
    axis.text.x = element_text(angle = -45, hjust = 0)
  )
```
